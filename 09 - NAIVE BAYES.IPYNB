## NAIVE BAYES

Es una forma de calcular la probabilidad de un evento basandose en el conocimiento previo de condiciones relacionadas con ese evento.
Es especialmente útil para problemas de calsificación (filtros)

# Supocisión "Naive"

La supocisión "Naive" (ingenuo) en Naive Bayes es que todas las caracteristicas son independientes entre sí dado la clase. Esto significa
que la presencia (o ausencia) de una caracteristica particular no está relacionada con la presencia (o ausencia) de cualquier otra caracteristica.
Esta supocisión simplifica en enormemente el cálculo

# PASOS PARA CLASIFICAR:
1. Calcular la probabilidad de priori (P(C))
2. Calcular la probabilidad condicional (P(F_i|C))
3. Calcular la probabilidad posterior
4. Seleccionar la clase con la mayor probabilidad posterior


```python
# importar bibliotecas
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
# Cargar el dataset
data = sns.load_dataset('iris')
# separacion de datos en Entrenamiento-Prueba
train, test = train_test_split(data, test_size= 0.4,
                               stratify=data['species'],
                               random_state=19)
```

```python
# Separación de la etiqueta de clase "Y" y atributos "XQ"
# Conjunto de entrenamiento
X_train = train[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y_train = train['species']
#Conjunto de pruebas
# Separación de la etiqueta de clase "Y" y atributos "XQ"

X_test = test[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y_test = test['species']
```

```python
# Clasficiador GAUSSIANO NAIVE BAYES
mod_gnb = GaussianNB()
y_pred = mod_gnb.fit(X_train, Y_train).predict(X_test)
print('La prediccion del Clasificador Gaussiano Naive Bayes en los datos de prueba es {:0.3f}'.format(metrics.accuracy_score(y_pred, Y_test)))
```

```python
pred2 = ['petal_length', 'petal_width']
mod_gnb_2pred = GaussianNB()
y_pred2 = mod_gnb_2pred.fit(X_train[pred2], Y_train).predict(X_test[pred2])
print('La prediccion del Clasificador Gaussiano Naive Bayes en los datos de prueba es {:0.3f}'.format(metrics.accuracy_score(y_pred2, Y_test)))
```

```python
test["prediccion"] = y_pred
test["prediccion2"] = y_pred2
test
```
